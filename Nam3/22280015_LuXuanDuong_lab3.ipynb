# Phần 2

### Sử dụng với Data chí có 2 loại trái cây

```python
df_new = df2[(df['fruit_name'] == "apple") | (df2['fruit_name'] == "orange")]
```

```python
df_new
```

#### Chia tập dữ liệu

```python
x= df_new.iloc[:,0:-1]
y= df_new.iloc[:,-1]
(x_train, x_test, y_train, y_test)= train_test_split(x, y, test_size = 0.3, random_state=42)
```

Dựa vào các Feature(có biến liên tục và biến rời rạc ) và sự gioiw thiệu về "Naive Bayse" trên Sklearn thì em thấy có 2 model có thể dùng để phân loại với bài này là
- GaussianNB
- Categorical NB
- Multinomial NB

### Xài GausianNB

```python
Gau = GaussianNB()
Gau.fit(x_train, y_train)
y_pred= (Gau.predict(x_test))
accuracy_score(y_pred, y_test)
```

```python
x= df_new.iloc[:,0:-2]
y= df_new.iloc[:,-1]
(x_train, x_test, y_train, y_test)= train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
Gau = GaussianNB()
Gau.fit(x_train, y_train)
y_pred= (Gau.predict(x_test))
accuracy_score(y_pred, y_test)
```

#### Xài MultinomialNB

```python
Mul = MultinomialNB()
Mul.fit(x_train, y_train)
```

```python
y_pred= (Mul.predict(x_test))
accuracy_score(y_pred, y_test)
```
