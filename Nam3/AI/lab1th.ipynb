```python
from collections import defaultdict
from queue import Queue, PriorityQueue

#doc data tu file.txt
def read_txt(file):
    size = int(file.readline())
    start, goal = [int(num) for num in file.readline().split(' ')]
    matrix = [[int(num) for num in line.split(' ')] for line in file]
    return size, start,goal, matrix
#Chuyen doi ma tran
def convert_graph(a):
    adjList= defaultdict(list)
    for i in range(len(a)):
        for j in range(len(a[i])):
            if a[i][j] == 1:
                adjList[i].append(j)
    return adjList

def convert_graph_weight(a):
    adjList= defaultdict(list)
    for i in range(len(a)):
        for j in range(len(a[i])):
            if a[i][j] != 0:
                adjList[i].append((j, a[i][j]))
    return adjList
```

# Cai dat BFS

```python
def BFS(graph, start, end):
    visited = []
    frontier = Queue()
    #Them node start vao frontier va visited
    frontier.put(start)
    visited.append(start)

    #start khong co node cha
    parent = dict()
    parent[start] = None
    path_found = False

    while True:
        if frontier.empty():
            print("No way Exception")  # Da duoc fix lai
            break
        current_node = frontier.get()
        # khong can thiet dung len "visited.append(current_node)"
        #visited.append(current_node)
        # Kiem tra current_node co la node end hay khong
        if current_node == end:
            path_found = True
            break

        for node in graph[current_node]:
            if node not in visited:
                frontier.put(node)
                parent[node] = current_node
                visited.append(node)

    path=[]
    if path_found:
        path.append(end)
        while parent[end] is not None:
            path.append(parent[end])
            end= parent[end]
        path.reverse()
    return path
    
```

# cai dat DFS

```python
def DFS(graph, start, end):
    visited = []
    frontier = []
    #Them node start vao frontier va visited
    frontier.append(start)
    visited.append(start)

    #start khong co node cha
    parent = dict()
    parent[start] = None

    path_found = False

    while True:
        if frontier==[]:
            print("No way Exception")  # Da duoc fix lai
            break
        current_node = frontier.pop()
        # khong can thiet dung len "visited.append(current_node)"
        # visited.append(current_node)
        # Kiem tra current_node co laf node end hay khong
        if current_node == end:
            path_found = True
            break

        for node in graph[current_node]:
            if node not in visited:
                frontier.append(node)
                parent[node] = current_node
                visited.append(node)

    path=[]
    if path_found:
        path.append(end)
        while parent[end] is not None:
            path.append(parent[end])
            end= parent[end]
        path.reverse()
    return path
    
```

# cai dat UCS

```python
def UCS(graph, start, end):
    visited = []
    frontier = PriorityQueue()
    #Them node start vao frontier va visited
    frontier.put((0,start))
    visited.append(start)
    cost_to_node = {start: 0} # cost o node start o dau la 0
    #start khong co node cha
    parent = dict()
    parent[start] = None
    path_found = False
    
    while True:
        if frontier.empty():
            print("No way Exception")  # Da duoc fix lai 
            break
        current_w, current_node = frontier.get()
        # Kiem tra current_node co laf node end hay khong
        if current_node == end:
            path_found = True
            break

        for nodei in graph[current_node]:
            node, weight = nodei
            cost = current_w + weight # tính toán chi phí
            if weight > 0:
                if node not in cost_to_node or cost < cost_to_node[node]: # so sánh chi phí 
                    cost_to_node[node] = cost
                    frontier.put((cost, node))
                    parent[node] = current_node
                    if node not in visited:
                        visited.append(node)

    path=[]
    if path_found:
        path.append(end)
        while parent[end] is not None:
            path.append(parent[end])
            end= parent[end]
        path.reverse()
    return current_w, path
    
```

```python
from collections import defaultdict, deque

class Graph:
    def __init__(self, file=None, weighted=False):
        self.file = file
        self.adjList = defaultdict(list)
        self.weighted = weighted
        if file:
            self.read_file()
    
    def read_file(self):
        with open(self.file, "r") as file:
            self.size = int(file.readline())
            self.start, self.goal = [int(num) for num in file.readline().split()]
            self.matrix = [[int(num) for num in line.split()] for line in file]

        # Convert matrix to adjacency list based on weight
        if self.weighted:
            self.convert_graph_weight()
        else:
            self.convert_graph()

    def convert_graph(self):
        self.adjList = defaultdict(list)
        for i in range(len(self.matrix)):
            for j in range(len(self.matrix[i])):
                if self.matrix[i][j] == 1:
                    self.adjList[i].append(j)

    def convert_graph_weight(self):
        self.adjList = defaultdict(list)
        for i in range(len(self.matrix)):
            for j in range(len(self.matrix[i])):
                if self.matrix[i][j] != 0:
                    self.adjList[i].append((j, self.matrix[i][j]))
  
    def BFS(self):
        visited = set()  # Use set for faster lookups
        frontier = deque([self.start])
        parent = {self.start: None}  # Keep track of the parent of each node
        path_found = False

        while frontier:
            current_node = frontier.popleft()

            if current_node == self.goal:
                path_found = True
                break

            if current_node not in visited:
                visited.add(current_node)
                for neighbor in self.adjList[current_node]:
                    if neighbor not in visited and neighbor not in frontier:
                        frontier.append(neighbor)
                        parent[neighbor] = current_node

        return self.reconstruct_path(parent) if path_found else []

    def DFS(self):
        visited = set()  # Use set for faster lookups
        frontier = [self.start]
        parent = {self.start: None}  # Keep track of the parent of each node
        path_found = False
        while frontier:
            current_node = frontier.pop()

            if current_node == self.goal:
                path_found = True
                break

            if current_node not in visited:
                visited.add(current_node)
                for neighbor in self.adjList[current_node]:
                    if neighbor not in visited:
                        frontier.append(neighbor)
                        parent[neighbor] = current_node

        return self.reconstruct_path(parent) if path_found else []
    
    def UCS(self):
        visited = []
        frontier = PriorityQueue()
        #Them node start vao frontier va visited
        frontier.put((0,self.start))
        visited.append(self.start)
        cost_to_node = {self.start: 0} # cost o node start o dau la 0
        #start khong co node cha
        parent = dict()
        parent[self.start] = None
        path_found = False
        
        while True:
            if frontier.empty():
                print("No way Exception")  # Da duoc fix lai 
                break
            current_w, current_node = frontier.get()
            # Kiem tra current_node co laf node end hay khong
            if current_node == self.goal:
                path_found = True
                break

            for nodei in self.adjList[current_node]:
                node, weight = nodei
                cost = current_w + weight # tính toán chi phí
                if weight > 0:
                    if node not in cost_to_node or cost < cost_to_node[node]: # so sánh chi phí 
                        cost_to_node[node] = cost
                        frontier.put((cost, node))
                        parent[node] = current_node
                        if node not in visited:
                            visited.append(node)

        return current_w, self.reconstruct_path(parent) if path_found else []      
    
    def reconstruct_path(self, parent):
        path = []
        current_node = self.goal
        while current_node is not None:
            path.append(current_node)
            current_node = parent[current_node]
        path.reverse()  # Reverse the path to get the correct order
        return path

    def display_path(self, bfs_path=None, dfs_path=None, ucs_cost=None, ucs_path=None, weight=None):
        self.weight=weight
        if (self.weight == False ):
            print("BFS Path:", bfs_path)
            print("DFS Path:", dfs_path)
        else:
            if ucs_path is not None:
                print("UCS Cost:", ucs_cost)
                print("UCS Path:", ucs_path)

```
