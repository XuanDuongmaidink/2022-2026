```python
# Em th·∫•y gi√° tr·ªã '2028-02 v·ª´a l√† gi√° tr·ªã c·ªßa outlier chi·∫øm t·ªâ l·ªá r·∫•t nh·ªè v√† v·ª´a kh√¥ng ƒë√∫ng th·ªùi gian giao  d·ªãch (Th·ªùi gian c·∫£u t∆∞∆°ng lai) n√™n em s·∫Ω drop n√≥
df = df[~(df['TransactionTime'] == '2028-02')]
```

```python
# H√†m ƒë·ªÉ l·ªçc outler b·∫±ng t·ª© ph√¢n v·ªã
def remove_outliers(df, column):
    out=[]
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    out= df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return out
```

Khi nh√¨n v√°o boxplot v√† histogram th√¨ em th·∫•y 2 c·ªôt 'NumberOfItemsPurchased' v√† 'CostPerItem' ƒëi·ªÅu c√≥ gi√° tr·ªã outlier l·ªõn nh∆∞ng l·∫°i chi·∫øm t·ªâ l·ªá nh·ªè n√™n khi drop s·∫Ω kh√¥ng ·∫£nh h∆∞·ªõng ƒë·∫øn d·ªØ li·ªáu

```python
outlier_NumberOfItemsPurchased= remove_outliers(df,'NumberOfItemsPurchased')
outlier_CostPerItem= remove_outliers(df,'CostPerItem')
# Em t·∫°o 2 bi·∫øn ƒë·ªÉ tr√°nh b·ªã sao khi drop outlier c·ªßa 1 c·ªôt th√¨ data s·∫Ω b·ªã ·∫£nh h∆∞·ªüng (lo·∫°i outlier ƒë·ªôc l√¢p)
```

```python
df = df[~df['CostPerItem'].isin(outlier_CostPerItem['CostPerItem'])]
df = df[~df['NumberOfItemsPurchased'].isin(outlier_NumberOfItemsPurchased['NumberOfItemsPurchased'])]
```
